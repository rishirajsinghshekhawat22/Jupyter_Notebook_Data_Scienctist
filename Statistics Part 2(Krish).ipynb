# Z Score

x = {1,2,3,4,5}
Mean = 3
SD = 1 (For example, we are considering this)
(I have plotted this histogram in notepad.)

Z Score = xi-mean/SD
= 1-3/1 = -2
= 2-3/1 = -1
= 3-3/1 = 0
= 4-3/1 = 1
= 5-3/1  = 2

After doing that, what we get is **Normal Distribution**. Thereafter, we will consider mean as 0 and SD as 1, then we will get Standard Normal Distribution.

For SND, Mean == 0, SD = 1 (Always)

Random Variable belongs to Standard Normal Distribution only when Mean = 0 and SD =1

There's a process called **Standardization**

Example: For say we have three data. (1st Application of Z-Score)
Age(Years), Weight(KG),Height(Cm)
Measurement units are different, hence we use **standarization** to bring all attributes to same scale.

After applying Z score, we will get the sync data. 

* Age-Age Mean/SD
* Height-Mean/SD
* Weight-Mean/SD

# Mutual Exclusive Event And Non Mutual Exclusive Event

Mutual Exclusive = Two events are mutually exclusive or disjoint if they cannot both occur at the same time.

Non Mutual Exclusive = If two events can both occur at the same time.

* Rolling a dice {Mutual Exclusive Event}
* Rolling two dices {Non Mutual Exclusive Event}
* Tossing a coin {Mutual Exclusive Event}
* Tossing two coins {Non Mutual Exclusive Event}
* Taking a card from a deck. Can we get king and queen? {Non Mutual Exclusive Event}

# Independent and Dependent events:
Example: Rolling of dice {1,2,3,4,5,6}
Pr(1) = 1/6
Pr(2) = 1/6
Pr(3) = 1/6
Pr(4) = 1/6

Above is the case of **independent event**, as each action won't effect the other one.


**Dependent Event**
In a bag of marbles there are various colors(Red, White, Black, Green, Pink)

Pr(White Marbel) = 1/5

Pr(Red Marbel) = 1/4 (Becuase after taking the White marbel, only 4 marbel were left)

It is dependent because it has dependency on other event.

Example: 
Rolling a dice(Independent Event). 


Rolling a dice. What is the probability of getting 1 and 6 in sequence.
In first throw, 1 should come and 6 in the second throw.

Pr(1 and 6) = Pr(1) * Pr(6)
= 1/6 * 1/6 = 1/3

Example: Dependent Event

A bag of marbles, we have 5 marbles with White, Redy, Blue, orange and Green 

What is the probability of taking out white marbel and then the red marbel.

Pr(White) = 1/5
Pr(Red) = 1/4

Pr (White and Red) = Pr(White) * Pr(R given white(Means after giving up white, we are left with 4 marbles)) It is also called conditional probability.
1/5*1/4 = 1/20

# Permutation And Combination 

Example: You went to a Zoo and have been given a task. 
In Zoo there are Tiger, Lion, Chimpanzee, Giraffe, Monkey, Snakes

A piece of paper has been given to you. List down the name of first three animals you see.

You get into the Zoo, you first sight a Tiger, 5 outcomes left, then you see Lion, 4 outcomes left.

* 6*5*4 = 120(The possible outcomes. This is permutation)
* Here, the order will matter. 
* At first you have seen Tiger, Lion And Chimpanzee
* Second You see Lion, Tiger, Chimanzee

**Permutation** Formula = npr = n factorial/(n-r)factorial

= 6*5*4*3factorial/3 factorial = 120


**Combination**

* Lion Tiger Chimpanzee
* Tiger Lion Chimpanzee

Different combination with same dataset will be treated as one. Like in the above case.

Combination Formula = ncr = n factorial/r factorial(n-r) Factorial
