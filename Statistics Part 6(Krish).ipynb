# Why sample variance is divided by n-1?

Population Mean                   Sample Mean

For an example, we have a 80 population data with a mean 500. However we are taking only 5 as a sample then there will be a huge difference between a sample and population mean and it will be called biased estimation, hence we use n-1 to get the closer mean to the population.

Scientist performed experiments. Whny it is only n-1 and not n-2 or n-3.

Estimate              Parameter

Sample Mean is estimate parameter which is a population mean.

When scientists divided sample and population by n. Sample mean was very less with population mean.
When they divided by n-1 mean was very close of both.
When they divided by n-2. Sample mean was > population mean.

Variance talk about distribution and spread. 
Mean is just a measure of central tendency, that's why we don't use n-1 for mean calculation. However, we can use if we want.


# Chi Square Test

**It is a non parametric test, that is performed on categorical data.**

**Interviewer will always ask why do you use this and that test?**

Example: 
In the 2000 USA census, the age of individual in a small town were found to be the following:

                    Percentage
* Less than 18            20%
* 18-35                   30%
* Greater than 35         50%

In 2010, age of n/sample = 500 individuals were sampled.
Below are the results:
                    Obeserved Values (F0)      Expected Value (F1) (500*%)
Less than 18                121                       100
18-35                       288                       150
Greater than 35             91                        250

Expected Value = Sample Size* Percentage. (500 * 20% (Means out of 500, 100 will be less than 500.))

Using alpha=0.05, can you conclude the distribution or ages has been changed in 10 years?

**Steps:**
1. Hypothesis
   1. H0 = The data meets the age distribution.(No changes in age.)
   2. H1 = Distribution of age has changed.

2. Alpha = 0.5, CI = 95%
3. Degree of freedom = Category-1, 3-1 = **2**
4. In chi square table, we get alpha and degree of freedom. We checked the table, it's 5.99
5. Decision rule:
   1. If Chi square value is > 5.99. we will reject the null hypothesis.
6. Calculate Chi Square Test
   chi square = summation of (frequency of observered-frequency of expected)2/Frequence of expected

Less than 18+18-35+>35 = (121-100)square/100 + (288-150)square/150 + (91-250)square/250
=232.94

We will reject the null hypothesis as chi square is >5.99.

We can conclude that the distribution of age has got changed.

**For degree of freedom we did 3-1 because here we focus on categorical values**

To accept the null hypothesis we were expecting the same value as last census but got something else. Hence to calculate expected value we used previous census data.

# Annova (F Test)

The t-test works well when dealing with two groups, but sometimes we want to compare more than two group at the same time. 

For example, if we wanted to test whether petal_width age differs based on some categorical variable like species, we have to compare the means of each level or group the variable.

One Way F-Test(Annova):-
It tells whether two or more groups are similar or not based on their mean similarity and f-score. 

Example: There are 3 different category of iris flowers and their petal width and need to check whether all 3 groups are similar or not.
