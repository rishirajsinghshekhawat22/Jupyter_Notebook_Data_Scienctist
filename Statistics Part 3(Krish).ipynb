# Covarriance
Let's assume we have a two types of data. 
* Weight(x) And Height(y)
* Quantify the relationship between x and y.
* Quantify the relation means, if you x value increases, y value increases OR x(down),y(down)

To find out this relation, we use covarriance.

There could be inverse or direct relationship. 

* X(up),y(up) = direct relation
* x(down),y(up) = Inverse relation

What is the purpose the Covarriance?

Quantifing the relationship between two independent feature. 

Formula = Summation of (x-x_bar)*(y-y_bar)/n-1

**Direct/Positive Relation**
x     y
2     5
3     6
4     7
5     8

* x_bar = X_mean = 14/4 = 3.5
* y_bar = Y_mean = 26/4 = 6.5

Calculation: 

(2-3.5)*(5-6.5)+(3-3.5)*(6-6.5)+(4-3.5)*(7-6.5)+(5-3.5)*(8-6.5)/3 

(2.25)+(0.25)+(0.25)+(2.25)/3
= 1.666

**Inverse Relation**
x     y
2     8
3     7
4     6
5     5

Calculation in notebook. 
* Mean_x = 3.5
* Mean_y = 6.5
* Covarriance = -1.666

* Positively correlated when output is positive.
* Negatively correlated when output is negative.
* Not correlated when output is zero.

Correlation can be done between dependent and dependent to independent as well.


**Disadvantages of Covarriance**

With Covarriance we can find the quantity relation but not how much directly it's correlated.

There's no limit, if the values are high the correlation will also high but we do not have any measures to calculate, how much directly it's correlated

We need to restrcit the outcome between -1 to 1.

Disadvantage of covarriance is taken care by **Pearson Correlation Coefficeint**.

# Spearman's Rank Correlation Coefficient.

Pearson correlation will not give us perfect 1 if there are slight changes upword on both axis. However, Spearman considers the slightest(More linear) of movement give 1 if both are increasing together.

In pearson if the points in the straight line then only it's 1. 
To capture the non linear, we use Spearman's Rank.

**Spearman Formula** = covarriance(rank of x, rank of y)/SD of rank of x and SD of rank of y.

**Example**

x     y      Rank of x                  Rank of y
5     100    Rank 2                         4
4     500    Rank 3                         2
7     200     Rank 1 7 is the highest       3
3     1000    Rank 4                        1

With the help of adding the attributes and apply that to the formulas, even the non linear movement will also be captured.

We are assiging ranks, to capture slightest of movement. Based on higher number we can rank data.
