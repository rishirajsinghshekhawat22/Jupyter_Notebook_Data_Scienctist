# Binomial Distribution
In Binomial we will have many Bernouli distribution.

An experiment is a Binomial experiment if :
1. It is repeated for fixed number of times.
2. The Trials are independent. (e.g., Tossing a coin)
3. Trials have 2 mutually exclusive outcomes, either success or failure.
4. The probability of success is same for all trials.(Unbiased experiments)

B(n,p)
* n = Number of trials
* p = success probability for each trial.

# Chebyshev's Inequality.

**If x is a Normal Disrtibution**

**Imperical Formula**

Probability (Mu - Sigma <= x <= Mu+Sigma) = 68% 
Probability (Mu - 2 Sigma <= x <= Mu- 2 Sigma) = 95%
Probability (Mu - 3 Sigma <= x <= Mu + 3 Sigma) = 99.7%

**What if we have another variable Y which is not a Gaussan Distribution. Can we find out how much % in the 1st, 2nd and 3rd SD?**

For this we use Chebyshev's Inequality.

Probability (Mu-K Sigma <= Y <= Mu+ K Sigma )>= 1-1/K square

K Sigma = Means 2 SD.

Condition : K value should be equal to or greater than 1.



If the distribution is not Gaussan then we can't find the data belong to which level or we can say, we can't apply Imperical Formula. Hence we use Chebyshev's Inequality.

**We will check whether a distribution is gussian distributed or not we use QQ Plot**

```python
### We will check whether a distribution is gussian distributed or not. 
### QQ Plot (Quantile Quantile plot)
import numpy as np
import statsmodels.api as sm
import pylab as py
```

```python
data_points= np.random.normal(0,1,100)
sm.qqplot(data_points,line='45')
py.show

# Blue dots which are falling in the straight line will make a Gaussian distribution.
```

```python
import seaborn as sns
sns.distplot(data_points)
```

# Log Normal Distribution

```python
# How to create log normal values?
# Create log normal distributed data.
gfc = np.random.lognormal(0,1,100)
sm.qqplot(np.log(gfc),line = '45')
py.show
```
