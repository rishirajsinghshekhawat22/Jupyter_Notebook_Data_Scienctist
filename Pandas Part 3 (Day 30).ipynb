```python
np.random.seed(23)
data =pd.DataFrame(np.random.randn(4,5),index = ['a','b','c','d'], columns = ['x','y','z','u','v'])
data
```

```python
# We have to extract negative value from the above data.

negative_data = data.values[data[::]<0]
list(negative_data)
```

```python
# Print column and row name of each dataset.

data

for i in data.columns:
    for j in data.index:
        if data[i][j] in negative_data:
            print(i,j,data[i][j])
        
```
