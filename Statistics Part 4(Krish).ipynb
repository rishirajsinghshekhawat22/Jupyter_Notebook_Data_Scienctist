# One Sample Z Test For Propotion.

A survey claims that 9 out of 10 doctors reccommend Aspiron for the patients with headache. To test this claim, a random sample of 100 doctors is obtaine. Out of these 100 doctors, 82 indicate that they recommend Aspiron.
Is this claim accurate? with Alpha 0.05

1. Null Hypothesis = P = 0.9 9/10
2. Alternate Hypothesis = P != 0.9

Calculation of Z statistcs for propotion.

P_Head = 0.82
P0 = 0.90
n=100

**Formula for Z test for propotion**
Z= P_head-P0/Root of (P0(1-P0)/n)

Z = -2.667
Reject Null Hypothesis
